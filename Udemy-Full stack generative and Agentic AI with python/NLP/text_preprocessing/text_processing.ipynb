```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import pandas as pd
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
df=pd.read_csv('tripadvisor-hotel-reviews.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df['Review_lower']=df['Review'].str.lower()
df.head(5)
```

```python
en_stopwords=stopwords.words('english')
en_stopwords.remove("not")
df['Review_lower']=df['Review_lower'].apply(lambda x: " ".join(word for word in x.split() if word not in en_stopwords))
```

```python
import re
df['Review_lower']=df.apply(lambda x: re.sub(r"[^\w\s]","",x['Review_lower']),axis=1)
```

```python
df['Review_lower'] = df.apply(
lambda x: word_tokenize(x['Review_lower']), axis=1)
```

```python
ps = PorterStemmer()
df["Review_lower"] = df["Review_lower"].apply(lambda tokens: [ps.stem(token) for
token in tokens])
```

```python
lemmatizer = WordNetLemmatizer()
df["Review_lower"] = df["Review_lower"].apply(lambda tokens:
[lemmatizer.lemmatize(token) for token in tokens])
```

```python
tokens_clean = sum(df['Review_lower'], [])
```

```python
unigrams = pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts()
```

```python
bigrams = pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts()
```

```python
ngrams_4 = pd.Series(nltk.ngrams(tokens_clean, 4)).value_counts()
```
