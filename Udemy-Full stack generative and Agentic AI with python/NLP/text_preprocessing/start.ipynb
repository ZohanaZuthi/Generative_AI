```python
reviews = [
    "The product quality is amazing and delivery was very fast.",
    "I am disappointed with the customer service experience.",
    "Absolutely loved it! I will buy again.",
    "The packaging was damaged but the item works fine.",
    "Very affordable and worth the money.",
    "The app interface is confusing and slow sometimes.",
    "Excellent performance and battery life.",
    "Not bad, but I expected better quality for the price.",
    "Customer support solved my issue quickly.",
    "Terrible experience, the product stopped working after two days.",
    "The design looks modern and professional.",
    "Shipping took longer than expected.",
    "I highly recommend this to my friends and family.",
    "The instructions were unclear and difficult to follow.",
    "Great value for money and easy to use."
]
```

sentence with look and s is optional

```python
pattern_to_find=r"looks?"
saras_review=[]
for str in reviews:
    if(re.search(pattern_to_find,str)):
        saras_review.append(str)
```

```python
print(saras_review)
```

starts with T use ^

```python
pattern_to_find=r"^T"
saras_review=[]
for str in reviews:
    if(re.search(pattern_to_find,str)):
        saras_review.append(str)
print(saras_review)
```

ended with e. use $

```python
pattern_to_find=r"e.$"
saras_review=[]
for str in reviews:
    if(re.search(pattern_to_find,str)):
        saras_review.append(str)
print(saras_review)
```

```python
pattern_to_find=r"(stop|expect)ed"
saras_review=[]
for str in reviews:
    if(re.search(pattern_to_find,str)):
        saras_review.append(str)
print(saras_review)
```

### Clean punctuation
here [] means set of words , inside [], ^ means not, \w means word charachter, \s means white space

```python
pattern_to_find=r"[^\w\s]"
saras_review=[]
for str in reviews:
    clean_review=re.sub(r"[^\w\s]","",str)
    saras_review.append(clean_review)
print(saras_review)
```

# Stemming

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for str in reviews:
    words=str.split()
    stemmed_words=[]
    for w in words:
        stemmed_words.append(ps.stem(w))
    print(stemmed_words)
    
    
```

the limitation of stemming is that it is not meaningful so lemmatization is better


# Lemmatization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
for str in reviews:
    words=str.split()
    stemmed_words=[]
    for w in words:
        stemmed_words.append(lemmatizer.lemmatize(w))
    print(stemmed_words)
```

# N Grams
N grams represents the relationship between words
Unigram -1 word
Bigram- 2 words
Trigram-3 words

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.util import ngrams
```

```python
sentence = "The product quality is amazing and delivery was very fast And delivery man was good The Product suits me"
words = word_tokenize(sentence.lower())
unigrams=pd.Series(ngrams(words,1)).value_counts()
print(unigrams)
unigrams[:10].sort_values().plot.barh(color='orange', figsize=(12,8))
```

```python
bigrams=pd.Series(ngrams(words,2)).value_counts()
print(bigrams)
bigrams[:10].sort_values().plot.barh(color='orange', figsize=(12,8))
```

```python
triigrams=pd.Series(ngrams(words,3)).value_counts()
print(triigrams)
triigrams[:10].sort_values().plot.barh(color='orange', figsize=(12,8))
```
