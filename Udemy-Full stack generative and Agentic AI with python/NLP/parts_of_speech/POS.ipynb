###  SpaCy is a library for Natural Language Processing (NLP). It allows you to analyze and understand text with Python.
tokenization, POS tagging, lemmatization, named entity recognition, dependency parsing

```python
import spacy 
import pandas as pd
```

python -m spacy download en_core_web_sm

loads a small English model that comes with vocabulary, grammar rules, and statistical 
patterns. 

```python
nlp=spacy.load('en_core_web_sm')
```

```python
text = "Apple is looking at buying a startup in London for $1 billion."
doc=nlp(text)
for token in doc:
    print(token.text,token.pos_,token.lemma_)
```

```python
df=pd.DataFrame(columns=['token','pos_tag'])
for token in doc:
    df=pd.concat([df,pd.DataFrame.from_records([{'token':token.text,'pos_tag':token.pos_}])],ignore_index=True)
```

```python
df.head(5)
```

```python
ts = (
    df.groupby(['token', 'pos_tag'])
      .size()
      .reset_index(name='counts')
      .sort_values(by='counts', ascending=False)
)
ts
```
